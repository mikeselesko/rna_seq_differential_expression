```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# file path
file_path = '../data/GSE102045_normalized_counts.csv'

# read in the matrix
df = pd.read_csv(file_path, index_col=0)

print(f"The shape of the dataframe is: {df.shape}")
df.head()
```

```python
df.isna().sum().sum()
df = df.apply(pd.to_numeric, errors='coerce')
```

## Data Exploration

```python
plt.hist(df.values.flatten(), bins=100)
plt.xlabel('Normalized expression level')
plt.ylabel('Frequency')
plt.title('Overall Expression Distribution')
plt.show()
```

```python
plt.hist(df.values.flatten(), bins=100, log=True)
plt.xlabel('Normalized expression level')
plt.ylabel('Log Frequency')
plt.title('Expression Distribution (Log Frequency)')
plt.show()
```

```python
# Log-transform the data to compress extreme outliers
log_counts = np.log2(df.values.flatten() + 1)

plt.hist(log_counts, bins=100)
plt.xlabel('log2(Normalized expression level + 1)')
plt.ylabel('Frequency')
plt.title('Log2-Transformed Expression Distribution')
plt.show()
```

```python
df.boxplot(figsize=(15, 5), rot=90)
plt.title('Sample-wise Expression Distribution')
plt.ylabel('Expression Level')
plt.show()
```

```python
df_log = np.log2(df + 1)

df_log.boxplot(figsize=(15, 5), rot=90)
plt.ylabel('log2(Normalized expression level + 1)')
plt.title('Sample-wise Expression Distribution (log2-transformed)')
plt.show()
```
